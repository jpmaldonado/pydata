### Command vs Edit mode
- If you press `ESC` you can switch to command mode.
- `ESC`+`A`: Insert a cell above
- `ESC` + `B`: Insert a cell below
- `ESC` + `dd`: Delete a cell
- `ESC` + `m`: Switch from code to markdown (text editor mode).
- `ESC` + `y`: Switch from markdown to code.
To execute a cell:
- `SHIFT` + `ENTER`: Runs and jumps to cell below.
- `CTRL` + `ENTER`: Runs a cell in place
- `ENTER`: Get inside a cell and modify its content.

## Filter by mask

- A **mask** is a set of True/False conditions that we use to filter our data.

```python
# From the original data frame, retrieve records that satisfy inner cond.
df[df['NUM_RATE_RSD']>20].head() 
```

```python
# M
df.loc[df['NUM_RATE_RSD']>20, ['DAT_RIF_M', 'IMP_RATE_RSD']].head()
```

```python
# More than one condition: Use () followed by & (AND) / | (OR)

df.loc[(df['NUM_RATE_RSD']>20) & (df['IMP_RATE_RSD']<7000), ['DAT_RIF_M', 'IMP_RATE_RSD']].head()
```
